## Get Data

### Minute data from Alpaca markets

using BTC/USD Data in this example

```python
base = "https://data.alpaca.markets/v2/stocks/bars?"
r1 = "symbols=SPY&timeframe=1T&start=2019-01-01&end=2023-12-31"

headers = {
    "accept": "application/json",
    "APCA-API-KEY-ID": "PKWEUWO01XKVEUS3QZT5",
    "APCA-API-SECRET-KEY": "7A8lCe62IdyvYOEbSw612DPbvolqgkvhTmQqUZhj"
}
```

```python
response = requests.get(base+r1, headers=headers)
```

```python
data_min = pd.DataFrame(response.json()['bars']['SPY'])

data_min['t'] = pd.to_datetime(data_min['t'], format='%Y-%m-%dT%H:%M:%S')

data_min['date'] = data_min['t'].dt.date
#data_min.set_index('t', inplace=True)
```

```python
# get all pages of data and merge exisiting dataframes with next page's dataframe
while response.json()['next_page_token']:
    page_token = response.json()['next_page_token']
    response = requests.get(base+r1+"&page_token="+page_token, headers=headers)
    next_data = pd.DataFrame(response.json()['bars']['SPY'])
    next_data['t'] = pd.to_datetime(next_data['t'], format='%Y-%m-%dT%H:%M:%S')
    #next_data.set_index('t', inplace=True)
    next_data['date'] = next_data['t'].dt.date
    data_min = pd.concat([data_min,next_data])
```
